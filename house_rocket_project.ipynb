# Questão de Negócio
A House Rocket é uma plataforma digital que tem como modelo de negócio, a compra e a venda de imóveis usando tecnologia. Sua principal estratégia é comprar boas casas em ótimas localizações com preços baixos e depois revendê-las posteriormente à preços mais altos. Quanto maior a diferença entre a compra e a venda, maior o lucro da empresa e portanto maior sua receita.

Entretanto, as casas possuem muitos atributos que as tornam mais ou menos atrativas aos compradores e vendedores e a localização e o período do ano também podem influenciar os preços. A seguir vamos responder as seguintes perguntas:

### 1. Quais são os imóveis que a House Rocket deveria comprar e por qual preço ?
### 2. Uma vez a casa comprada, qual o melhor momento para vendê-las e por qual preço ?

## Repondendo a primeira questão de negócio:
1. Quais são os imóveis que a House Rocket deveria comprar?

```python
#descobrindo a mediana de preço por região
median_per_zipcode = data[['zipcode',
                           'price']].groupby('zipcode').median().reset_index()
median_per_zipcode.columns = ['zipcode', 'median price']
display(median_per_zipcode)
```

```python
# juntando os dataframes
df = pd.merge(data, median_per_zipcode, how='outer')
```

```python
# imóveis que deveria comprar ==> imóveis com valor menor que a mediana da região, condição >=3 e nota >= 8
df = df[['id', 'price', 'zipcode', 'grade', 'median price', 'condition']]
best_houses = df[(df['price'] < df['median price']) & 
                 (df['condition'] >= 3) &
                 (df['grade'] >= 8)]
best_houses
```

## Repondendo a segunda questão de negócio:
2. Uma vez a casa comprada, qual o melhor momento para vendê-las e por qual preço ?

```python
#descobrindo a melhor época do ano para vender imóvel

avg_price_per_month = data[['month', 'price']].groupby('month').mean('price').reset_index()
sns.lineplot(data=avg_price_per_month,
              x='month',
              y='price')
```

```python
#descobrindo a mediana do preço por região e período do ano
colunas = data[['month', 'zipcode', 'price']].copy()
month_zipcode = colunas.groupby(['zipcode', 'month']).median().reset_index()
month_zipcode.columns = ['zipcode', 'month', 'reference price']

#considerando apenas os melhores meses para venda (4 - 6)
month_zipcode = month_zipcode[(month_zipcode["month"] >= 4)
                              & (month_zipcode['month'] <= 6)]
month_zipcode = month_zipcode.groupby('zipcode').mean().reset_index()
month_zipcode = month_zipcode.drop(columns=['month'])

month_zipcode.head()  #descobrimos a referência de preço para cada região no melhor período de vendas do ano
```

```python
best_houses = pd.merge(best_houses, month_zipcode, how='outer')
best_houses.head(20)
```

```python
best_houses['sale price'] = 'NA'

for i in range(len(best_houses)):
    if best_houses.loc[i, 'price'] > best_houses.loc[i, 'reference price']:
        best_houses.loc[i, 'sale price'] = 1.1 * best_houses.loc[i, 'price']

    elif best_houses.loc[i, 'price'] <= best_houses.loc[i, 'reference price']:
        best_houses.loc[i, 'sale price'] = 1.3 * best_houses.loc[i, 'price']
```
